## Feature Selection

The challenge to this analysis was to find the measurement that could indicate the performances of batting, pitching and fielding. As a reminder from the previous analysis, Moneyball analysis, the goal of baseball teams is to win. To win, you need runs and to gain runs you need to get on base. Based on this concept, I will be using Base Earned as the indicator to measure the performance of each skills. 


### Batting

From the data, Sean 'Lahman' Baseball Database, we can get the base earned by batting using the following equation:

$TB(Total Base) = 1B + (2*2B) + (3*3B) + (4*HR)$

$Base_{batting} = TB + BB$

Then we would want to weight the numbers with PA(Plate Appearance), to capture the opportunity difference between teams. 
    

```python
b = bat.groupby(['teamID','yearID']).sum().reset_index()
b['PA'] = b.AB + b.BB + b.IBB + b.HBP + b.SH + b.SF
b['1B'] = b.H - b['2B'] - b['3B'] - b.HR
b['bat'] = (b['1B'] + (2*b['2B']) + (3*b['3B']) + (4*b.HR) + b.HBP + b.BB) / b.PA

# merge it to team df
team = pd.merge(team, b[['teamID', 'yearID','PA','bat','1B']], how='left', on=['teamID','yearID'])
```

### Pitching

As for pitching, there is a given equation by Bill James to estimate the pitcher's total bases(PTB) based on number of hits and walks allowed. The equation goes by the following:

$BaseAllowed_{pitching} = 0.89 * (1.255(H - HR) + 4*HR) + 0.56 * (BB + HBP - IBB)$

This equation only uses Homeruns allowed, hit batter and walks. Base on the theory of Defense Independent Pitching Statistics(DIPS), these stats are almost entirely determined by pure pitcher's ability. 

Then the equation will by weighted by BF(Batters Faced), which is basically PA for pitchers.

```python
p = pitch.groupby(['teamID','yearID']).sum().reset_index()
p['pitch'] = -((0.89 * (1.255 * (p.H - p.HR) + (4*p.HR)) + 0.56 * (p.BB + p.HBP - p.IBB)) / p.BFP)

# merge it to team df
team = pd.merge(team, p[['teamID', 'yearID','pitch']], how='left', on=['teamID','yearID'])
```

### Fielding

For fielding, we will be using errors(E), double plays(DP), opponent stolen base(OSB) and opponent caught stealing(OCS). 

$Base_{fielding} = (OCS + PO + DP) - (OSB + E)$

We will make an assumption that for every errors, a batter is put on base. So that we can give a minus to each given base to opponent and plus to each base saved. Then will be weighted by PA.

```python
f = field.groupby(['teamID','yearID']).sum().reset_index()
f['field'] = f.CS + f.DP + f.PO - f.SB - f.E

# merge it to team df
team = pd.merge(team, f[['teamID', 'yearID','field']], how='left', on=['teamID','yearID'])
team['field'] = team.field / team.PA
```

### Base Running

As for base stealing, we can use stats such as, caught stealing and stolen base, which seems to be the only base running related stats to calculate total base by stealing. 

$Base_{stealing} = SB - CS$

Remember that, with limited data, we are still missing some base running related stats. There are more to base running other than SB and CS.

```python
b['steal'] = (b.SB - b.CS) / b.PA

# merge it to team df
team = pd.merge(team, b[['teamID', 'yearID','steal']], how='left', on=['teamID','yearID'])
```

```python
#Additional Stats
team = team.fillna(0)
```

```python
team = pd.merge(team, weight, how='left', left_on='yearID', right_on='Season')
team = pd.merge(team, p[['teamID','yearID','HBP']], how='left', on=['teamID','yearID'])

team['wOBAt'] = ((team.wBB*team.BB) + (team.wHBP*team.HBP_x) + (team.w1B*team['1B']) + (team.w2B*team['2B']) + (team.w3B*team['3B']) + (team.wHR*team.HR)) / (team.AB+team.BB+team.SF+team.HBP_x)
team['FIP'] = ((team.HRA*13 + (team.BBA+team.HBP_y)*3 - team.SOA*2)) / team.IPouts + team.cFIP
team['SO/BB'] = team.SO / team.BB
```

## Historical Era Comparison

MLB has endured many changes in rules, equipment, and strategies over time. Because of such shifts, Major League Baseball has been segmented into distinct eras. For this reason, will have to see if there are significant difference between each eras.

1) **The Dead Ball Era(1901~1919**

  During the era, there were some changes in rule that favored pitchers. First, the size og the home plate got bigger along with the strike zone. Second, foul balls began to count as strikes.
  
2) **The Live Ball Era(1920~1941)**

  Pitchers were no longer permitted to alter baseball or use tricks like spitball and emergy ball. Umpires were assigned to put new balls into play whenever the ball became scuffed or discolored.
  
3) **The Integration Era(1942~1960)**

  In this era, a lot of African American players were integrated to the MLB.

1) **The Expansion Era(1961~1976)**

  First, American League joined the MLB which increased the teams from 16 to 24 teams. Second, the strike zone was extended **from** *the armpits to the top of the knees* **to** *shoulders to the bottom of the knees*. Third, a new position, Designated Hitters(DH), was created to hit instead of the pitcher.

2) **The Free Agency Era(1977~1993)**

  The first year of this era, 24 players became free agents and the average salary for the player was six times what it had been. The abundant team got stronger.

3) **The Steroid Era(1994~2005)**
 
  During this time, the use of steroid was populated and the MLB did not strictly check for steroids. 

4) **The Post-Steroid Era(2006~)**

  Beginning with the 2006 season, the penalties for testing positive for steroids became harsher. Player were banned for 50 games for the first offense, 100 games for the second and lifetime ban for the next. With harsher penalties, it helped players deter the use of steroids. 

for more information about baseball eras click [here](https://thesportjournal.org/article/examining-perceptions-of-baseballs-eras/)

```python
# era
for n in range(team.shape[0]): 
    if team.loc[n,'yearID'] >= 1901 and team.loc[n,'yearID'] <= 1919:
        team.loc[n,'era'] = 'Deadball'
    if team.loc[n,'yearID'] >= 1920 and team.loc[n,'yearID'] <= 1941:
        team.loc[n,'era'] = 'Liveball'
    if team.loc[n,'yearID'] >= 1942 and team.loc[n,'yearID'] <= 1960:
        team.loc[n,'era'] = 'Integration'
    if team.loc[n,'yearID'] >= 1961 and team.loc[n,'yearID'] <= 1976:
        team.loc[n,'era'] = 'Expansion'
    if team.loc[n,'yearID'] >= 1977 and team.loc[n,'yearID'] <= 1993:
        team.loc[n,'era'] = 'Free-agency'
    if team.loc[n,'yearID'] >= 1994 and team.loc[n,'yearID'] <= 2005:
        team.loc[n,'era'] = 'Steroid'
    if team.loc[n,'yearID'] >= 2005:
        team.loc[n,'era'] = 'Post-steroid'
```

### One-way ANOVA Test

**ANOVA Assumptions:**
1) Each sample is randomly selected and independent

2) Homoscedasticity

3) The experimental errors are normally distributed

```python
# Homoscedasticity Assumption
# Levene's test
# batting
result = []
for val in ['bat','pitch','field','steal']:
    d = team[team.era == 'Deadball'][val]
    l = team[team.era == 'Liveball'][val]
    i = team[team.era == 'Integration'][val]
    e = team[team.era == 'Expansion'][val]
    f = team[team.era == 'Free-agency'][val]
    s= team[team.era == 'Steroid'][val]
    p = team[team.era == 'Post-steroid'][val]
    result.append(levene(d,l,i,e,f,s,p))

pd.DataFrame(result, index=['Bat','Pitch','Field','Run'])
```

We used **Levene's test** to check if the data have equal variance. 
The null and alternative hypotheses of Leven's test are:

$ H_0 = \sigma_1^2 = \sigma_2^2 = \sigma_3^2 $

$ H_1 = \sigma_i^2 \neq \sigma_j^2 $

We can see that Pitching was the only resulted with p-value of greater than 0.05, which means that it have strong probability of having equal variance. However, as for Batting, Fielding and Base running, we could not null hypothesis was rejected, which means that we have to use **Welch ANOVA**.

```python
# ANOVA for Batting and Pitching
model_pitch = ols('pitch ~ C(era)', team).fit()
anova_test = anova_lm(model_pitch)
anova_test.index = ['Pitching','Residual']
```

```python
# Welch's ANOVA for Fielding and Running
model_bat = ols('bat ~ C(era)', team).fit()
model_field = ols('field ~ C(era)', team).fit()
model_steal = ols('steal ~ C(era)', team).fit()
welch_bat = welch_anova(dv='bat', between='era', data=team)
welch_field = welch_anova(dv='field', between='era', data=team)
welch_run = welch_anova(dv='steal', between='era', data=team)
welch_test = pd.concat([welch_bat, welch_field, welch_run])
welch_test.index = ['Batting','Fielding','Running']
print('-----------------------------one-way ANOVA-----------------------------\n')
print(anova_test)
print('\n\n-----------------------------Welch\'s ANOVA-----------------------------\n')
print(welch_test)
```

The null and alternative hypotheses for **one-way ANOVA test** and **Welch's ANOVA** goes by the following:


$ H_0 = \mu_i = \mu$ for all $i$

$ H_1 = \mu_i \neq \mu$ for some $i$

According to the test above, the p-value for all four variables is approximately 0, thus we reject $H_0$ and have strong evidence that there's at least one pairwise group mean difference in each stats at a 5% significant level.

```python
# Normality Assumption
f,ax = plt.subplots(2,2, figsize=(20,20))
plt.suptitle('Probability plot of residuals')
normality_plot, stat = stats.probplot(model_bat.resid, plot= ax[0,0], rvalue= True)
normality_plot, stat = stats.probplot(model_pitch.resid, plot= ax[0,1], rvalue= True)
normality_plot, stat = stats.probplot(model_field.resid, plot= ax[1,0], rvalue= True)
normality_plot, stat = stats.probplot(model_steal.resid, plot= ax[1,1], rvalue= True)
plt.show()
```

Now that we know there is at least one pairwise group mean difference in each variables. But which of the eras are significantly different? 


To find out which specific groups have statistically significance difference in mean, we will perform a post-hoc test with **Tukey's test** for ANOVA and **Games Howell's test** for Welch's ANOVA.

```python
# post-hoc test
comp_bat = mc.MultiComparison(team.bat, team.era)
tukey_bat = comp_bat.tukeyhsd()
comp_pitch = mc.MultiComparison(team.pitch, team.era)
tukey_pitch = comp_pitch.tukeyhsd()
comp_field = mc.MultiComparison(team.field, team.era)
tukey_field = comp_field.tukeyhsd()
comp_run = mc.MultiComparison(team.steal, team.era)
tukey_run = comp_run.tukeyhsd()
print('-----------------------------Batting-----------------------------\n')
print(pg.pairwise_gameshowell(dv='bat', between='era', data=team)[['A','B','diff','pval']])
print('\n\n-----------------------------Pitching-----------------------------\n')
print(tukey_pitch.summary())
print('\n\n-----------------------------Fielding-----------------------------\n')
print(pg.pairwise_gameshowell(dv='field', between='era', data=team)[['A','B','diff','pval']])
print('\n\n-----------------------------Running-----------------------------\n')
print(pg.pairwise_gameshowell(dv='steal', between='era', data=team)[['A','B','diff','pval']])
```

The post hoc test result table above for **batting and pitching** shows that almost all of the p-value is less than 0.05 except the group mean differences of **Live Ball era, Integration era and Free-agency era**. Other than these 3 groups, the pairwise group mean differences are **statistically significant** at a 5% significance level.

While, the result table for **fielding** indicates that **other than Deadball and Liveball era, and Expansion and Integration era** the difference between groups are statistically significant. And at last, Base running result table shows that other than **Free-agency era, Steroid era and Post-Steroid era** difference between are statistically significant. 

## Variable Importance with Boruta

Now we will see the importance of each skills by the winning percentage for each eras.We will be using Boruta which is an algorithm that uses randomization on top of results obtained from variable importance obtained from randomforest to determine the truly important and statistically valid results. 

The problem with this impurity based randomforest is that it have a strong tendency to **overestimate** the importance of **continuous numerical** or **high cardinality categorical** variables. Thus, we will use Boruta to support our selection.

The basic idea of Boruta is to create a copy of variable, "shadow", for each variables on the dataset and shuffle them. Then merge the shadow variables with the original ones and make **comparison of original variables with the randomised variables** to measure the importance. The variables having higher importance than the randomised variables will be the ones that are considered to be relevant. 

```python
team = team.dropna(axis=1)
X = team[['bat','pitch','field','steal']]
y = team['Wp']

# split train and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# regression model
rf = RandomForestRegressor(n_jobs=-1, random_state=0)
cv = KFold(n_splits=5, random_state=0, shuffle=True)
n_estimators = np.arange(10,500,10)

r = []
for n in n_estimators:
    rf.set_params(n_estimators=n)
    scores = cross_val_score(rf, X_train, y_train, cv=cv, n_jobs=-1,)
    r.append([n,scores.mean()])

result = pd.DataFrame(r, columns=['n_estimators','score'])

plt.figure(figsize=(10,5))
plt.plot(result.n_estimators, result.score)
```

```python
result[result.score == result.score.max()].n_estimators
```

```python
n_estimators = result[result.score == result.score.max()].n_estimators.values
rf = RandomForestRegressor(n_jobs=-1, n_estimators=170, random_state=42)
rf.fit(X_train, y_train)
pred = rf.predict(X_test)

score = rf.score(X_test, y_test)
```

```python
print('R squared score:',score)
```

```python
rf = RandomForestRegressor(n_jobs=-1, n_estimators=590, random_state=42)
```

```python
# Boruta
X = X.values
y = y.values
selector = BorutaPy(rf, verbose=2, random_state=0)
# find all relevant features
selector.fit(X, y)
```

```python
# zip my names, ranks, and decisions in a single iterable
feature_ranks = list(zip(['Battng','Pitching','Fielding','Running'],
                         selector.ranking_, 
                         selector.support_))

# iterate through and print out the results
for feat in feature_ranks:
    print('Feature: {:<25} Rank: {},  Keep: {}'.format(feat[0], feat[1], feat[2]))
    
```

```python
eras = team.era.unique()
selected = []
for era in eras:
    df = team[team.era == era]
    X = df[['bat','pitch','field','steal']]
    y = df.Wp
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
    
    # model
    rf = RandomForestRegressor(n_jobs=-1, random_state=0)
    cv = KFold(n_splits=5, random_state=0, shuffle=True)
    n_estimators = np.arange(10,500,10)
    
    r = []
    for n in n_estimators:
        rf.set_params(n_estimators=n)
        scores = cross_val_score(rf, X_train, y_train, cv=cv, n_jobs=-1)
        r.append([n,scores.mean().astype('float')])
    
    result = pd.DataFrame(r, columns=['N','score'])
    n_best = result[result.score==result.score.max()].N.item()
    
    rf = RandomForestRegressor(n_estimators=n_best, n_jobs=-1, random_state=0)
    rf.fit(X_train, y_train)
    score = rf.score(X_test, y_test)
    
    print(f'\n\n-----------------------{era} era-----------------------')
    print('R squared:', score)
    
    importance = rf.feature_importances_
    selected.append(importance)
```

```python
result = pd.DataFrame(selected,
                      columns = ['Batting','Pitching','Fielding','Running'],
                      index = ['Deadball','Liveball','Integration','Expansion','Free-agency','Steroid','Post-steroid'])
```

```python
# plt.barh(['wOBA', 'FIP', 'FP'], result, stacked=True)
ax = result.plot.barh(stacked=True, figsize=(10,4), color=['royalblue','cornflowerblue','lightsteelblue','slategrey'])
plt.gca().invert_yaxis() # reverse order(top to down)
plt.title('Bat vs Pitch vs Field')

for n in ax.patches:
    width, height = n.get_width(), n.get_height()
    x, y = n.get_xy()
    ax.text(x+width/2,
            y+height/2,
            '{:.0f} %'.format(width*100),
            horizontalalignment='center',
            verticalalignment='center')
plt.savefig('Importance.png')
plt.show()
```

For all, batting and pitching have been the most significant skill for winning. It seems that in the past, batting has given more impact on teams winning percentage but since the Steroid era, pitching has got more important.
